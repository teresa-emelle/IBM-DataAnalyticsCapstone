# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
import numpy as np
QUERY = """
SELECT *
FROM master
"""
df = pd.read_sql_query(QUERY, conn)
df.head()
```

```python
df['ConvertedComp'].plot(kind='hist')
plt.title('Histogram of Converted Compensation')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
df['Age'].plot(kind='box')
plt.title('Box Plot of Age of Respondents')
plt.ylabel('Age of Respondents')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
df.plot(kind='scatter', x='Age', y='WorkWeekHrs')
plt.title('Relationship between Age and Work Week Hours')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
normalized_age = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
df.plot(kind='scatter', x='WorkWeekHrs',y='CodeRevHrs',s=normalized_age*1000,c='green',alpha=.2)
plt.title('Relationship betweek Work Week Hours, Code Rev Hours and Age')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Rev Hours')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT COUNT(DatabaseDesireNextYear) as Count, DatabaseDesireNextYear
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY COUNT(DatabaseDesireNextYear) DESC LIMIT 5
"""
df2 = pd.read_sql_query(QUERY,conn)
df2.set_index('DatabaseDesireNextYear', inplace=True)
df2
```

```python
df2.plot(kind='pie',autopct='%1.1f%%', figsize=(8.5,8.5), subplots=True, labels=None)
plt.title('Top 5 Databases Respondents want to Learn Next Year')
plt.legend(labels=df2.index, loc='upper right')
plt.show()
```

```python
QUERY = """
SELECT COUNT(LanguageDesireNextYear) as Count, LanguageDesireNextYear
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY COUNT(LanguageDesireNextYear) DESC LIMIT 5
"""
df3 = pd.read_sql_query(QUERY,conn)
df3.set_index('LanguageDesireNextYear', inplace=True)
df3
```

```python
QUERY = """
SELECT COUNT(LanguageWorkedWith) as Count, LanguageWorkedWith
FROM LanguageWorkedWith
GROUP BY LanguageWorkedWith
ORDER BY COUNT(LanguageWorkedWith) DESC LIMIT 5
"""
df4 = pd.read_sql_query(QUERY,conn)
df4.set_index('LanguageWorkedWith', inplace=True)
df4
```

```python
QUERY="""
SELECT DatabaseWorkedWith,COUNT(Respondent)
FROM DatabaseWorkedWith
GROUP BY Respondent HAVING COUNT(DatabaseWorkedWith)=1 AND DatabaseWorkedWith='MySQL'"""
df5=pd.read_sql_query(QUERY,conn)
df5
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age 
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df1 = pd.read_sql_query(QUERY,conn)
order = ['WorkWeekHrs', 'CodeRevHrs']
df1.groupby('Age')[order].median().plot.bar(stacked=True)
plt.title('Median Work Week Hours and Code Rev Hours for Ages 30-35')
plt.show()
```

### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp, Age 
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df6 = pd.read_sql_query(QUERY,conn)
order = ['ConvertedComp']
df6.groupby('Age')[order].median().plot.line(figsize=(20,6))
plt.title('Converted Compensation for Respondents Age 45-60')
plt.show()
```

```python
QUERY = """
SELECT ConvertedComp, Age 
FROM master
WHERE Age BETWEEN 25 AND 30
"""
df8 = pd.read_sql_query(QUERY,conn)
order = ['ConvertedComp']
df8.groupby('Age')[order].median().plot.line(figsize=(20,6))
plt.title('Converted Compensation for Respondents Age 25-30')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """

SELECT COUNT(MainBranch) as Count, MainBranch 
FROM master
GROUP BY MainBranch

"""

df7=pd.read_sql_query(QUERY,conn)

df7.set_index('MainBranch',inplace=True)
df7.plot(kind='barh', color='lightgreen')
plt.title('Horizantal Bar Chart of Main Branch')
plt.show()
```
